###Consolidar datos

```python
df_completo =pd.concat([tienda1,tienda2,tienda3,tienda4])
df_completo
```

```python
df_completo.info()
```

###Formato de fecha

```python
#fecha a datetime
df_completo['Fecha de Compra'] = pd.to_datetime(df_completo['Fecha de Compra'])
#valor total facturado
df_completo['Total_Facturado'] = df_completo['Precio'] + df_completo['Costo de envío']
```

```python
df_completo.info()
```

```python
df_completo.sample(5)
```

#1. Análisis de facturación



```python
facturacion_global=df_completo['Total_Facturado'].sum()
facturacion_global
```

```python
facturacion_tienda = df_completo.groupby('Tienda')['Total_Facturado'].sum()
```

```python
print("Facturación total por tienda:")
print(facturacion_tienda)
```

```python
#Comparamos la facturacion de las tiendas
import matplotlib.pyplot as plt
facturacion_tienda.plot(kind='bar', figsize=(10, 6))
plt.title('Comparación de Facturación por Tienda')
plt.xlabel('Tienda')
plt.ylabel('Facturación Total')
plt.xticks(rotation=45)
plt.grid(axis='y', linestyle='--')
plt.tight_layout()
plt.show()
```

# 2. Ventas por categoría

```python

ventas_categoria = df_completo.groupby('Categoría del Producto')['Total_Facturado'].sum().reset_index()
ventas_categoria


```

```python
import plotly.express as px

#gráfico de árbol
fig = px.treemap(ventas_categoria,
                 path=[px.Constant("Todas las Categorías"), 'Categoría del Producto'],
                 values='Total_Facturado',
                 title='Facturación por Categoría de Producto',
                 color_continuous_scale='Blues')
fig.show()
```

# 3. Calificación promedio de la tienda


```python
promedio_calificacion =df_completo.groupby('Tienda')['Calificación'].mean().reset_index()
promedio_calificacion
```

# 4. Productos más y menos vendidos

```python
#analisis de  ventas productos
ventas_por_producto = df_completo['Producto'].value_counts()


productos_mas_vendidos = ventas_por_producto.head(7)
productos_menos_vendidos = ventas_por_producto.tail(7)


print("productos más vendidos:")
print(productos_mas_vendidos)
print("productos menos vendidos:")
print(productos_menos_vendidos)
```

```python
plt.figure(figsize=(12, 6))
productos_mas_vendidos.plot(kind='barh', color='skyblue')
plt.title('Los 7 Productos Más Vendidos')
plt.xlabel('Cantidad de Ventas')
plt.ylabel('Producto')
plt.gca().invert_yaxis()
plt.show()
```

```python
plt.figure(figsize=(12, 6))
productos_menos_vendidos.plot(kind='barh', color='skyblue')
plt.title('Los 7 Productos menos Vendidos')
plt.xlabel('Cantidad de Ventas')
plt.ylabel('Producto')
plt.show()
```
